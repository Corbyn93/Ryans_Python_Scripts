# Spectrum analyser script

Here is a script to analyse the results of measurements recorded using the Aim TTi spectrum analyser. 

```python
import numpy as np 
import scipy.optimize as opt
import matplotlib.pyplot as plot 
import pandas as pd
```

```python
def import_data(file_name):
    '''Call in the data-file and save it as a pandas array.'''
    
    data = pd.read_table(file_name, skiprows = 17, names = ['Frequnecy (MHz)', 'Power (dBm)'], delimiter = ',')
    return(data)
```

```python
def set_plot_params():
    '''Set the plot parameters to be readable.'''
    
    plot.rcParams.update({'font.size' : 16})
    plot.rcParams.update({'figure.figsize' :[12, 8]})
    
```
