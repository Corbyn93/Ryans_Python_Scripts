# Plot resonant scanner images. 

The aim of this script is to create images showing the resonant scanner scan range. 

```python
import numpy as np
import matplotlib.pyplot as plot
from scipy.signal import argrelmin
```

```python
def import_image(file_name):
    
    image = np.loadtxt(file_name)
    return(image)
```

```python
def set_plot_params():
    # formats the plot parameters 
    plot.rcParams.update({'font.size' : 16})
    plot.rcParams.update({'figure.figsize' :[12, 8]})
```

```python
folder = 'H:\\My Documents\\Strathclyde_Back_up\\Thesis and Planning\\Thesis_plots\\Resonant_scanner\\'
file = '20200908_01'

set_plot_params()

image = import_image(folder + file + '.txt')
fig, ax = plot.subplots()

plot.imshow(image, cmap = 'gray')
plot.colorbar(shrink=0.9, label ='\nPixel Value')

x1 = [250, 1500]
y1 = [700, 700]

y2 = [660, 660]

plot.plot(x1, y1, color = 'r', linestyle = '--', linewidth = '2')
plot.plot(x1, y2, color = 'r', linestyle = '--', linewidth = '2')
plot.axis([0, 1600, 1200, 0])

plot.xlabel('Pixels')
plot.ylabel('Pixels')

#fig.savefig('H:\\My Documents\\Strathclyde_Back_up\\Thesis and Planning\\Thesis_plots\\Resonant_scanner\\' + 'res_scanner_calib_still' + '.pdf')
```
