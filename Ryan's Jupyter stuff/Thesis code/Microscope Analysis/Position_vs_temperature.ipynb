```python
import numpy as np 
import matplotlib.pyplot as plot
import scipy.optimize as opt
```

```python
plot.rcParams.update({'font.size' : 20})
plot.rcParams.update({'figure.figsize' :[12, 8]})
```

```python
def linear(x, m, c):
    y = m*x + c
    return(y)
```

```python
x = np.array([7.7, 5.3, 3.0, 0.65, -1.7])
y = np.array([4.5, 0.7, -2.5, -5.8, -9.8])
z = np.array([128.3, 121.9, 116.1, 111.5, 105.0])

temp_oko = np.array([30, 32, 34, 36, 38])
temp_therm = [31.3, 33.1, 34.8, 36.3, 38.5]

fig, ax = plot.subplots()
ax.scatter(temp_oko, x, color = 'k', marker = 'x')
plot.xlabel(r'Temperature ($^\circ$C)')
plot.ylabel(r'x position ($\mu$m)')
p0 = [0,0]
y_fit, con = opt.curve_fit(linear, temp_oko, x, p0)
y = linear(temp_oko, y_fit[0], y_fit[1])
plot.plot(temp_oko, y, linestyle = '--', color = 'k',
         label = 'y = '+ str(np.round(y_fit[0], 3)) + 'x + ' + str(np.round(y_fit[1], 3) ) ) 
plot.legend()


fig1, ax1 = plot.subplots()
ax1.scatter(temp_oko, y, color = 'k', marker = 'x')
plot.xlabel(r'Temperature ($^\circ$C)')
plot.ylabel(r'y position ($\mu$m)')
p0 = [0,0]
y_fit1, con1 = opt.curve_fit(linear, temp_oko, y, p0)
y1 = linear(temp_oko, y_fit1[0], y_fit1[1])
plot.plot(temp_oko, y1, linestyle = '--', color = 'k',
         label = 'y = '+ str(np.round(y_fit1[0], 3)) + 'x + ' + str(np.round(y_fit1[1], 3) ) ) 
plot.legend()


fig2, ax2 = plot.subplots()
ax2.scatter(temp_oko, z, color = 'k', marker = 'x')
plot.xlabel(r'Temperature ($^\circ$C)')
plot.ylabel(r'z position ($\mu$m)')
p0 = [0,0]
y_fit2, con2 = opt.curve_fit(linear, temp_oko, z, p0)
y2 = linear(temp_oko, y_fit2[0], y_fit2[1])
plot.plot(temp_oko, y2, linestyle = '--', color = 'k',
         label = 'y = '+ str(np.round(y_fit2[0], 3)) + 'x + ' + str(np.round(y_fit2[1], 3) ) ) 
plot.legend()
```
