# Non-reference CW ODMR Thermometry

Here I have created a script for analysing the full CW non-reference thermometry experiments performed in 2020 using the Oko-Labs environment chamber. 

```python
import numpy as np 
import scipy.optimize as opt 
import matplotlib.pyplot as plot 
import pandas as pd
```

```python
def import_data(file):
    
    data = pd.read_table(file)
    return(data)
```

```python
def set_plot_params():
    # formats the plot parameters 
    plot.rcParams.update({'font.size' : 20})
    plot.rcParams.update({'figure.figsize' :[12, 8]})
```

```python
def linear(x, m, c):
    y = m*x + c
    return(y)
```

```python
folder = 'H:\\My Documents\\Strathclyde_Back_up\\Thesis and Planning\\Thesis_plots\\Coplanar Waveguides\\'
date = '20200916'
file = '_reference_ODMR_analysis'

set_plot_params()

data = import_data(folder + date + '\\' + date + file + '.dat')

res_freq_1 = data['Res Freq 1 (GHz)']
error_1 = data['error res 1 (GHz)']

res_freq_2 = data['Res Freq 2 (GHz)']
error_2 = data['error res 2 (GHz)']

mean_res_freq = (res_freq_1 + res_freq_2)/2

tot_error = (np.sqrt(error_1**2 + error_2**2))

#print(mean_res_freq)

t1 = (mean_res_freq[0] + mean_res_freq[1])/2
t2 = (mean_res_freq[2] + mean_res_freq[3] + mean_res_freq[4])/3
t3 = (mean_res_freq[5] + mean_res_freq[6] + mean_res_freq[7])/3
t4 = (mean_res_freq[8] + mean_res_freq[9])/2
t5 = (mean_res_freq[10] + mean_res_freq[11] + mean_res_freq[12])/3
t6 = (mean_res_freq[13] + mean_res_freq[14] + mean_res_freq[15])/3

t_array = np.array([t1, t2, t3, t4, t5, t6])
t = np.array([27, 30, 32, 34, 36, 38])
e1 = np.sqrt(tot_error[0]**2 + tot_error[1]**2)
e2 = np.sqrt(tot_error[2]**2 + tot_error[3]**2 + tot_error[4]**2)
e3 = np.sqrt(tot_error[5]**2 + tot_error[6]**2 + tot_error[7]**2)
e4 = np.sqrt(tot_error[8]**2 + tot_error[9]**2)
e5 = np.sqrt(tot_error[10]**2 + tot_error[11]**2 + tot_error[12]**2)
e6 = np.sqrt(tot_error[13]**2 + tot_error[14]**2 + tot_error[15]**2)

error = np.array([e1, e2, e3, e4, e5, e6])
p = [0,0]
linear_params, conf = opt.curve_fit(linear, t, t_array, p, sigma = error)
linear_error = np.sqrt(np.diag(conf))

y = linear(t, linear_params[0], linear_params[1])

print(linear_params)
print(linear_error)

plot.errorbar(t, t_array, error, linestyle = '', color = 'k', 
              marker = 'x', label = 'Experiment Data')
plot.plot(t, y, color = 'r', label = 'Linear fit')
plot.legend()
plot.xlabel(r'Temperature ($^\circ$C)')
plot.ylabel('Mean Resonant Frequency (GHz)')

#plot.savefig('H:\\My Documents\\Strathclyde_Back_up\\Thesis and Planning\\Thesis_plots\\Coplanar Waveguides\\20200916\\' + '20200916_T_sensing.pdf')
```
