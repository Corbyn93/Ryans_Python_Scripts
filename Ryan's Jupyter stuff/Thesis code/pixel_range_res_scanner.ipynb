# Plotting the resonant_scanner line_scans. 

just a simple script to read in some data and generate a plot for the thesis. 

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plot
```

```python
def import_data(file_name):
    
    data = pd.read_table(file_name)
    return(data)
```

```python
def set_plot_params():
    # formats the plot parameters 
    plot.rcParams.update({'font.size' : 16})
    plot.rcParams.update({'figure.figsize' :[12, 8]})
```

```python
folder_path = 'H:\\My Documents\\Strathclyde_Back_up\\Thesis and Planning\\Thesis_plots\\Resonant_scanner\\'

file_name = 'calibrate_scan_angle_image.txt'
set_plot_params()

data = import_data(folder_path + file_name)
pixel = data['Distance_(pixels)']
brightness = data['Gray_Value']
```

```python
fig, ax = plot.subplots()
plot.plot(pixel, brightness, color = 'k')
plot.xlabel('Pixel Number')
plot.ylabel('Grey Value')

x1 = [195, 195]
y1 = [0, 1400]

x2 = [830, 830]
y2 = [0, 1400]

plot.plot(x1, y1, linestyle = ':', color = 'k')
plot.plot(x2, y2, linestyle = ':', color = 'k')

plot.annotate('', xy=(195, 400), xytext=(830, 400), arrowprops=dict(arrowstyle='<->'))
plot.text(480, 420, r'$\Delta~$Pixels')
plot.axis([0, 1050, 0, 1450])
```
