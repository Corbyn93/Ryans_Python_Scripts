```python
import numpy as np 
import matplotlib.pyplot as plot
```

```python
voltage = [4.5, 4, 3.5, 3, 2.5, 2, 1.5, 1, 0.75, 0.5, 0.25, 0.001]
peak = [4.76, 4.32, 3.76, 3.16, 2.72, 2.16, 1.72, 1.22, 0.96, 0.69, 0.38, 0.12]
```

```python
fig, ax = plot.subplots()
ax.scatter(voltage, peak, marker = 'x', color = 'k', label ='data point')
x = np.linspace(0,5, 100)
y = 1.0271*x + 0.10

plot.rcParams.update({'font.size' : 16})
plot.rcParams.update({'figure.figsize' :[12, 8]})
plot.plot(x, y, color = 'r', label = '1.02x + 0.1')
plot.xlabel('DC applied Voltage (V)')
plot.ylabel('Optical scan angle (deg)')
plot.legend()
```
