This code is designed to load in the resonant frequency vs temperature data file that has been created from the analysis of the NVScan_analysis.dat file.
The data file is read into a pandas array and then 3 graphs are made. The first two plot how the strain split peaks of the NV centre vary as temperature is increased. The third one rapresents the shift of the resonant frequnecy of the NV centre with temperature if there were no applied magnetic fields or strain splitting. This "Central frequency" is determined from the mean of the resonant frequencies of the strain split peaks. 

A linear fit is then made to the data (or in the first instance a portion of this data) to determine the change of resoanant frequnecy with temperature for the nanodiamond in question. 

The errors associated with these measurements (at least when first created) are derived from the standard deviation of multiple measurements that have been made on the same nanodiamond at each temperature step. 

The plots are then exported and saved as .png and .pdf files in the appropriate data folder within the users H drive. 

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plot 
import scipy.optimize as opt
```

```python
def get_data(folder_path, file_name):
    # Import the data file data into a Pandas array. 
    file_path = folder_path + file_name
    
    data_frame = pd.read_table(file_path)
    
    return(data_frame)
```

```python
def plot_f_split_vs_T(experiment_data):
    # This method plots the strain split frequencies vs temperature on differnet subplots. 
    
    fig, (ax1, ax2) = plot.subplots(2,1)
    plot.rcParams.update({'font.size' : 16})
    plot.rcParams.update({'figure.figsize' :[8, 12]})
    
    Temp = experiment_data['Stage Temperature (oC)']
    low_freq_resonance = experiment_data['Resonant Frequnecy 1 (GHz)']
    error_low_freq_resonance = experiment_data['error Resonant Frequency 1 (GHz)']
    high_freq_resonance = experiment_data['Resonant Frequnecy 2 (GHz)']
    error_high_freq_resonance = experiment_data['error Resonant Frequency 2 (GHz)']
    ax1.errorbar(Temp, low_freq_resonance, error_low_freq_resonance, linestyle = '', marker = 'x',
                label = 'Experiment data, low frequency dip')
    ax1.set_xlabel('Temperature (oC)')
    ax1.set_ylabel('Resonant frequnecy (GHz)')
    ax2.errorbar(Temp, high_freq_resonance, error_high_freq_resonance, linestyle = '', marker = 'x',
                label = 'Experiment data, low frequency dip')
    ax2.set_xlabel('Temperature (oC)')
    ax2.set_ylabel('Resonant frequnecy (GHz)')
    
    return(fig, ax1, ax2)
```

```python
def linear_graph(x, m, c):
    # creates a linear fit
    y = m*x + c
    return(y)
```

```python
def get_r_squared(x, y_data, y_fitting):
    # This is a function used to generate the goodness of fit metric r^2.
    # This value will be used to judge the effectiveness of the curve fitting 
    # used to find the resonant frequencies of the NV centre.
    
    y_data_mean = np.mean(y_data)
    SStot = np.sum((y_data - y_data_mean)**2) # Find the variation in experimental y values.
    SSres = np.sum((y_data - y_fitting)**2) # Find the variation betweem the experimental and fitted y values. 
    
    r_squared = 1 - (SSres/SStot) # Calculate R^2
    
    return(r_squared)
```

```python
#Get the data file path and pass it to the get_data() method. 
folder_path = 'H:\\My Documents\\Strathclyde_Back_up\\Graphs_and_analysis\\Thermometry\\20200916\\'
file_name = '20200916_average_res_freq_vs_temp.dat'
experiment_data = get_data(folder_path, file_name)

# unpack data frame 

Temp = experiment_data['Stage Temperature (oC)']
low_freq_resonance = experiment_data['Resonant Frequnecy 1 (GHz)']
error_low_freq_resonance = experiment_data['error Resonant Frequency 1 (GHz)']
high_freq_resonance = experiment_data['Resonant Frequnecy 2 (GHz)']
error_high_freq_resonance = experiment_data['error Resonant Frequency 2 (GHz)']
central_resonance_freq = experiment_data['Central Resonant Frequnecy (GHz)']
error_central_resonance_freq = experiment_data['mean Central resonant frequency error (GHz)']

# plot the strain split resonant frequencies 
#fig1, ax1, ax2 = plot_f_split_vs_T(experiment_data)

# plot the central frequnecy vs temperature data. 
fig, ax3 = plot.subplots()
ax3.errorbar(Temp, central_resonance_freq, error_central_resonance_freq, 
            linestyle = '', marker = 'x', color = 'k',
            label = 'Experimental data')
ax3.set_xlabel('Temperature (oC)')
ax3.set_ylabel('Resonant frequency (GHz)')
plot.rcParams.update({'font.size' : 16})
plot.rcParams.update({'figure.figsize' :[10, 10]})

# fitting a stright line to the central frequnecy data 
p0 = [1,1]
coeffs, confidence = opt.curve_fit(linear_graph, Temp, central_resonance_freq, P0)
fitted_data = linear_graph(Temp, coeffs[0], coeffs[1])
ax3.plot(Temp, fitted_data, label = 'Straight line fit. \nGradient = ' + str(np.round(coeffs[0]*1000, 3)) + ' MHz/K',
        color = 'r')
ax3.legend()
r_squared = get_r_squared(Temp, central_resonance_freq, fitted_data)
print('R_Squared = ' + str(round(r_squared,3)))
plot.savefig(folder_path + '20200916_central_freq_vs_T_plot.png')
plot.savefig(folder_path + '20200916_central_freq_vs_T_plot.pdf')
```
