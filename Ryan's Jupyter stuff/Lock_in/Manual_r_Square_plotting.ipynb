```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plot 
```

```python
def read_data(file):
    #get data into pandas array
    df = pd.read_table(file)
    return(df)
```

```python
 def set_plot_params():
    # formats the plot parameters 
    plot.rcParams.update({'font.size' : 20})
    plot.rcParams.update({'figure.figsize' :[12, 8]})
```

```python
folder = 'I:\\Science\\Physics-Nanobiophotonics\\Group\\User Data Analysis\\Ryans Data (2018-2021)\\ODMR\\20201125\\Lock_in_adaptive\\additional_analysis\\'
file = '20201125_NVScan_double_peak_analysis'

data = read_data(folder+file+'.dat')

r_squared = data['R^2 for fitting'].to_numpy()
rms_error = data['RMS error for fitting (normalised data)'].to_numpy()

r_squared = np.reshape(r_squared,[12, 5])
rms_error = np.reshape(rms_error, [12,5])

line_scans = np.array([10, 20, 30, 40, 50])
lock_in_freqs = ['500', '500', '400', '400', '200', '200']
dwell_time = [200, 400, 250, 500, 500, 1000]

fig, ax = plot.subplots()
fig2, ax2 = plot.subplots()
set_plot_params()
for i in range(6):
    ax.cla() # or ax.clear()
    ax2.cla()
    print(i*2)
    print((i*2)+1)
    ax.scatter(line_scans, r_squared[i*2,:], color = 'r', 
              label='Lock-in ODMR data', marker = 'o')
    ax.scatter(line_scans, r_squared[(i*2)+1,:], color = 'k', 
              label='CW ODMR data', marker = 'x')
    ax.set_title('R Squared vs line scans, AOM Mod freq = ' + 
                lock_in_freqs[i] + ' Hz, \nPixel dwell time = ' + 
                str(dwell_time[i]) + r' $\mu$s' ', Microwave Dwell time = '+
                 str(20000) + r' $\mu$s')
    ax.set_xlabel('Number of Line Scans')
    ax.set_ylabel('R squared')
    ax.legend(loc='lower right')
    
    ax2.scatter(line_scans, rms_error[i,:], color = 'r', 
              label='Lock-in ODMR data', marker = 'o')
    ax2.scatter(line_scans, rms_error[(i*2)+1,:], color = 'k', 
              label='CW ODMR data', marker = 'x')
    ax2.set_title('RMS Error vs line scans AOM Mod freq = ' + 
                lock_in_freqs[i] + ' Hz, \nPixel dwell time = ' + 
                str(dwell_time[i]) + r' $\mu$s' ', Microwave Dwell time = '+
                 str(20000) + r' $\mu$s')
    ax2.set_xlabel('Number of Line Scans')
    ax2.set_ylabel('RMS Error')
    ax2.legend(loc='upper right')

    #fig.savefig(folder + file + lock_in_freqs[i] + 'Hz' + str(dwell_time[i]) + 'us_R_Sqaured' + '.png')
    #fig2.savefig(folder + file + lock_in_freqs[i] + 'Hz' + str(dwell_time[i]) + 'us_RMS_Error' + '.png')
```
