```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plot
import os 
```

```python
def get_data(file_loaction):
    # get the data from the data file and put it in a panda dataframe. 
    data = pd.read_table(file_loaction)
    return(data)
```

```python
def create_contrast_plot(data, indicies, x_data_name):
    # 
    
    # formats the plot parameters 
    plot.rcParams.update({'font.size' : 20})
    plot.rcParams.update({'figure.figsize' :[12, 8]})
    
    
    fig1, ax1 = plot.subplots()
    ax1.scatter(data[x_data_name][indicies], data['Resonant frequency 1 contrast (%)'][indicies], 
                marker = 'x', color = 'r', label = 'Low freq resonance contrast', linestyle = '-')
    
    ax1.scatter(data[x_data_name][indicies], data['Resonant frequency2 contrast (%)'][indicies], 
                marker = 'o', color = 'b', label = 'High Freq resonance contrast')
    ax1.set_xlabel('Number of line scans')
    ax1.set_ylabel('ODMR Contrast (%)')
    ax1.legend(loc = 'lower left')
    
    return(fig1, ax1)
```

```python
def create_r_squared_plot(data, indicies, x_data_name):
    
    fig2, ax2 = plot.subplots()
    ax2.scatter(data[x_data_name][indicies], data['R^2 for fitting'][indicies], 
                marker = 'x', color = 'k', label = 'R^2 value for fit')
    
    ax2.set_xlabel('Number of line scans')
    ax2.set_ylabel('R_Squared value (A.U.)')
    ax2.legend(loc = 'lower right')
    return(fig2, ax2)
```

```python
def create_FWHM_plot(data, indicies, x_data_name):
    
    fig3, ax3 = plot.subplots()
    ax3.scatter(data[x_data_name][indicies], data['FWHM 1 (MHz)'][indicies], 
                marker = 'x', color = 'r', label = 'Low freq resonance FWHM')
    
    ax3.scatter(data[x_data_name][indicies], data['FWHM 2 (MHz)'][indicies], 
                marker = 'o', color = 'b', label = 'High freq resonance FWHM')
    ax3.set_xlabel('Number of line scans')
    ax3.set_ylabel('FWHM (MHz)')
    ax3.legend(loc = 'upper right')

    return(fig3, ax3)
```

```python
def save_plots(fig1, fig2, fig3, x_data_tag, date):
    
    save_folder = 'H:\\My Documents\\Strathclyde_Back_up\\Graphs_and_analysis\\ODMR\\20201023\\data_analysis\\'
    
    fig1.savefig(save_folder + date+'_contrast_vs_' + x_data_tag + '.png')
    fig1.savefig(save_folder + date+'_contrast_vs_' + x_data_tag + '.pdf')
    
    fig2.savefig(save_folder + date+'_r^2_vs_' + x_data_tag + '.png')
    fig2.savefig(save_folder + date+'_r^2_vs_' + x_data_tag + '.pdf')
    
    fig3.savefig(save_folder + date+'_FWHM_vs_' + x_data_tag + '.png')
    fig3.savefig(save_folder + date+'_FWHM_vs_' + x_data_tag + '.pdf')
    
```

```python
folder = 'I:\\Science\\Physics-Nanobiophotonics\\Group\\User Data Analysis\\Ryans Data (2018-2021)\\ODMR\\20210430\\power_sweep\\'
date = '20210430'
file = '20210430_NVScan_double_peak_analysis'

data = get_data(folder + date + '\\' + file + '.dat')

laser_test_index = np.arange(1,9,1)
dwell_time_index = np.arange(9,16,1)
averages_index = np.arange(16,24,1)

test = 'Averages'

fig1, ax1 = create_contrast_plot(data, averages_index, test)
fig2, ax2 = create_r_squared_plot(data, averages_index, test)
fig3, ax3 = create_FWHM_plot(data, averages_index, test)

save_plots(fig1, fig2, fig3, test, date)
```
